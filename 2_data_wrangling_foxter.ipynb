# Data Wrangling

The purpose of this notebook is to clean the data collected from Foxter Real State company website. This is the second step into the project to estimate prices of apartaments in Porto Alegre (Brazil).


### Summary
This notebook will execute the actions below:
1. Drop unnecessary columns
2. Convert columns to right formats
3. Verify and clean columns
4. Identify repeated elements
5. Handling missing values - NaNs
6. Remove outliers
7. Save the final dataframe to .csv file

### Import modules

```python
import numpy as np
import time
import pandas as pd
#import plotly.plotly as py
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

# outliers section
import seaborn as sns
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
import plotly.graph_objs as go

import warnings
warnings.filterwarnings("always")
```

### Load dataset

```python
df = pd.read_csv('2019-07-29-foxter.csv', sep='\t', decimal=',') # web_scraper notebook saves with date in the name of the file.

print("Total of rows:", df.shape[0])
```

```python
df.head()
```

```python
df.info()
```

```python

'''
Columns identification:

Unnamed - item identification
id - item identification
price - apartment price in Brazilian Real (float)
area - in m2 (float)
district - neighborhood (string)
city - city name (string)
type - kind of real state (house, apartment, etc.) (string)
segment - segment kind of real state (commercial, residential, etc.) (string)
condominium - price of condominium (float)
iptu - value of iptu tax (float)
rooms - rooms quantity (int)
box - quantity of parking space (int)
url - url of the real state (string)
date - date when collected (date)
'''

print()
```

___
### Drop unnecessary columns

```python
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
df.drop(['url'], axis = 1, inplace = True)
df.drop(['date'], axis = 1, inplace = True)

df.info()
```

___
### Convert columns to right formats

```python
df['area'] = df['area'].astype(float)
df['box'] = pd.to_numeric(df['box'], downcast='integer', errors='coerce')

df.info()
```

___
### Handling missing values - NaNs

The final option was to eliminate all missing values because is very dificult to estimate with high precision the "price" (dependent variable) and other independent variables as "area", "condominium" and "iptu".

```python
'''
df['price'].fillna(df['price'].mean(), inplace=True)
df['area'].fillna(df['area'].mode()[0], inplace=True)
df['segment'].fillna(df['segment'].mode()[0], inplace=True)
df['condominium'].fillna(df['condominium'].mode()[0], inplace=True)
df['iptu'].fillna(df['iptu'].median(), inplace=True)
'''

# Remove all NaNs
df = df.dropna()

df.info()
```

___
### Verify and clean "segment"

Our first interest is with "residencial" type.

```python
# identificar e retirar outros segmentos que não sejam o residencial
df.groupby(['segment']).count()
```

```python
df['segment'] = df['segment'].replace("Comercial, Residencial", "Residencial")
df.groupby(['segment']).count()
```

___
### Verify and clean "apartamentos"

We need to certify to mantain only apartment kind into the analysis.

```python
df.groupby(['type']).count()
```

```python
df = df.drop(df.query('(type != "Apartamento")').index)
df.groupby(['type']).count()
```

___
### Verify and clean "rooms"

Certify if rooms are described correctly.

```python
df.groupby(['rooms']).count()
```

```python
df['rooms'] = pd.to_numeric(df['rooms'], downcast='integer', errors='coerce')
df['rooms'].fillna(df['rooms'].mode()[0], inplace=True)
df.groupby(['rooms']).count()
```

___
### Verify "box" - garage

Certify if garages numbers are identified correctly.

```python
df['box'].value_counts(dropna=False)
```

___
### Maintaing "Porto Alegre" only

Maintain into the analysis only "Porto Alegre" city.

```python
df.groupby(['city']).count()
```

```python
df = df.drop(df.query('(city != "Porto Alegre")').index)
df.groupby(['city']).count()
```

___
### Identify repeated elements

```python
print ("total of elements: " + str(df.shape[0]))
print ("total of unique elements: " + str(df.id.nunique()))
print ("repeated elements: " + str(df.shape[0] - df.id.nunique()))
```

```python
# drop duplicates
df = df.drop_duplicates(subset=['id'])
print ("duplicated elements: " + str(df.shape[0] - df.id.nunique()))
```

```python
# identify duplicated

_ = df[df.duplicated(['id'], keep=False)]['id']
_.sort_values()
```

```python
# identify missing values

_ = df.isnull().sum(axis=0) 
print ("Missing values quantity: "+str(_.sum()))
```

___
### Identify and correct districts names

Verify if districts are spelled correctly.

```python
df['district'].value_counts().sort_index()
```

```python
df['district'] = df['district'].replace("Mont'Serrat", "Mont Serrat")
df['district'] = df['district'].replace('Humaita', "Humaitá")
df['district'] = df['district'].replace('Jardim Itu Sabará', "Jardim Itu-Sabará")
df['district'] = df['district'].replace('Jardim Lindoia', "Jardim Lindóia")
df['district'] = df['district'].replace('Jardim Sao Pedro', "Jardim São Pedro")
df['district'] = df['district'].replace('Passo da Areia', "Passo d'Areia")
df['district'] = df['district'].replace('Santa Cecilia', "Santa Cecília")
df['district'] = df['district'].replace('Santa Teresa', "Santa Tereza")
df['district'] = df['district'].replace('hipica', "Hípica")
df['district'] = df['district'].replace('Hipica', "Hípica")
df['district'] = df['district'].replace('Centro', "Centro Histórico")

df['district'].value_counts().sort_index()
```

___
### Maintaing only specified districts

Into the whole project, we are interested with 3 distincts districts.

```python
df = df.drop(df.query("(district != 'Mont Serrat' & district != 'Auxiliadora' & district != 'Bela Vista')").index)

df.groupby(['district']).count()
```

___
### Drop unnecessary columns

These columns will not necessary into the future analysis.

```python
df.drop(['id'], axis = 1, inplace = True)
df.drop(['city'], axis = 1, inplace = True)
df.drop(['type'], axis = 1, inplace = True)
df.drop(['segment'], axis = 1, inplace = True)

df.info()
```

___
### Outliers identification

The functions below will identify and remove outliers.

```python
# boxplot to identify outliers in price.

x = df.district

trace0 = go.Box(
    y=df.price,
    x=x,
    name='kale',
    marker=dict(
        color='blue'
    )
)

data = [trace0]
layout = go.Layout(
    yaxis=dict(
        title='price',
        zeroline=False
    ),
    boxmode='group'
)

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
# to find values into the boxplot
# https://stackoverflow.com/questions/33518472/getting-data-of-a-boxplot-pandas?rq=1

_, bp = pd.DataFrame.boxplot(df['price'], return_type='both')

outliers = [flier.get_ydata() for flier in bp["fliers"]]
boxes = [box.get_ydata() for box in bp["boxes"]]
medians = [median.get_ydata() for median in bp["medians"]]
whiskers = [whiskers.get_ydata() for whiskers in bp["whiskers"]]

min_outliers = outliers[0].min()
print ("Limit for cutting outliers:", min_outliers)

# remove outliers
df = df.loc[df['price'] < min_outliers]
```
